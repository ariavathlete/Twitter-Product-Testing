The casino game of craps is played as follows:
1. The player rolls a pair of standard 6-sided dice and takes their sum.
    a. If the sum is 7 or 11, then the player wins and the game is over .
    b. If the sum is 2, 3, or 12, then the player loses (this is called “crapping out”) and the game is over.
    c. If the sum is anything else, then we record the sum (lets call it “X”) and continue to the next step.


2. The player then rerolls the dice and takes their sum
    a. If the sum is X, the player wins and the game is over
    b. If the sum is 7, the player loses and the game is over
    c. If the sum is anything else, repeat step

#### 4. If you play the game of craps with these two dice, you will get one dollar if you win, and lose one dollar if you lose, then what is the expected return for playing the game?

```python
def craps():
    n, invest=get_input()
    cash = invest
    wins, cash = roll_sim(n, invest)
    Results(n, wins, cash, invest)
    return wins

def get_input():
    invest = eval(input("How much do you want to invest: $"))
    n=eval(input("Input the max amount of games of craps to simulate: "))
    return n, invest


def roll_sim(n, cash):
    roll_Count=0
    wins=0
    points=0
    result = ""
    
    

    while roll_Count < n and cash > 0:
            
            points = 0
            roll_Count=roll_Count+1

            playing=randrange(1,7) + randrange(2,5)

            if playing == 7 or playing == 11:
                wins = wins + 1
                cash = cash + 1
                result = "You Won Congrats!"
            
            elif playing == 2 or playing == 3 or playing == 12:
                wins = wins + 0
                cash = cash - 1
                result = "You Lost Sorry"

            else:
                
                result = "continue your game please"
#                 currentpoint = playing
                points = randrange(1,7) + randrange(2,5)

            while result == "continue your game please":

                if points == playing:
                    wins=wins+1
                    cash = cash + 1

                elif points == 7:
                    wins=wins
                    cash = cash - 1
                    
                break
                    
                
                
    return wins, cash


def Results(n, wins, cash, invest):
    
    returned =  100*(cash - invest)/invest
    
    print('\nFor', n, 'games of craps simulated,', wins, 
          'games were wins, giving a success rate of '+ str(100*(wins/n)) + '%.')
    print('Cash left is $', cash)
    print('Percentage returned is', str((returned/n)), ' %')


if __name__ == '__main__': craps()
# Input 100000 for number of simulations and amount to enter
```
